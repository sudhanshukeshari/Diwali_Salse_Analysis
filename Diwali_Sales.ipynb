```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load excel dataset
df = pd.read_excel('/content/Diwali Sales Data.excel.xlsx')
df
```

```python

df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop unnacessory columns
df.drop(['Status','unnamed1'], axis = 1, inplace=True)
```

```python
df
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
#Changing data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
Gen_count= sns.countplot(x='Gender', data= df)

for i in Gen_count.containers:
  Gen_count.bar_label(i)
```

```python
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data= sales_gen)
```

```python
From above graph we can observe that most of the buyers are Females and even most of the puchasing power of females are greater than males.
```

```python
age_count= sns.countplot(x='Age Group', hue= 'Gender', data= df )


for i in age_count.containers:
  age_count.bar_label(i)
```

```python
Age_Gr= df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.barplot(x='Age Group', y='Amount', data=Age_Gr)
```

 **From above graph we can see that most of the buyer and purchare are of age group between 25-35 are female**

```python
#Total number of order within the states

state_orders= df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x='State', y='Orders', data= state_orders)
```

```python
# Total number amount within the states
state_amount= df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x='State', y='Amount', data= state_amount)
```

**From above graph we can observe that maximum number of orders and maximum number of buyers are from uttar pradesh state**

**Marital Status**

```python
ax= sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
for i in ax.containers:
  ax.bar_label(i)
```

```python
marital_st= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status', y='Amount', hue= 'Gender', data=marital_st)
```

**From above graph we can see that most of the buyers are married and have most purchsing power**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(x='Occupation', data=df)

for i in ax.containers:
  ax.bar_label(i)
```

```python
sns.set(rc={'figure.figsize':(20,5)})

occupation= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Occupation', y='Amount', data=occupation)
```

** From above graph we can see that most of the buyers and purchaser comes from IT Sector, Healthcare and Aviation**

```python
sns.set(rc={'figure.figsize':(25,5)})
ax= sns.countplot(x='Product_Category', data=df)

for i in ax.containers:
  ax.bar_label(i)
```

**From above graph we can see that clothing & Apparel , Food, Electronics & Gadgets have higher sales than other products **

```python
sns.set(rc={'figure.figsize':(30,5)})

product= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Product_Category', y='Amount', data=product)
```

**From above graph we can see that most of the peole spend their money on food, clothing and apparel, electronic**
